```python
import pandas as pd
import numpy as np
data = pd.read_csv('/Users/suki/Desktop/valve.csv')
data_target = data['Target']

y = np.where(data_target == 0,0,1)
to_drop = ['Target']
churn_feat_space = data.drop(to_drop,axis=1)
churn_feat_space
X = churn_feat_space.as_matrix().astype(np.float)
```

```python
#For same hidden layer(1) with different neurons
```

```python
from sklearn.learning_curve import validation_curve
from sklearn.neural_network import MLPClassifier

def rms_error(model, X, y):
    y_pred = model.predict(X)
    return np.sqrt(np.mean((y - y_pred) ** 2))

hidden_layer_sizes = [(10,),(50,),(100,),(200,),(300,),(400,),(500,),(600,),(700,),(800,)
                      ,(900,),(1000,),(1200,),(1400,),(1600,),(1800,),(2000,),(2200,)]

train_scores, valid_scores = validation_curve(MLPClassifier(), X, y, 'hidden_layer_sizes',
                                              hidden_layer_sizes, scoring=rms_error)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

def plot_with_err(x, data, **kwargs):
    mu, std = data.mean(1), data.std(1)
    lines = plt.plot(x, mu, '-', **kwargs)
    plt.fill_between(x, mu - std, mu + std, edgecolor='none',
                     facecolor=lines[0].get_color(), alpha=0.2)

list = [10,50,100,200,300,400,500,600,700,800,900,1000,1200,1400,1600,1800,2000,2200]
plot_with_err(list, train_scores, label='training scores')
plot_with_err(list, valid_scores, label='validation scores')
plt.xlabel('hidden_layer_sizes'); plt.ylabel('rms error')
plt.legend();
```

```python
from sklearn.learning_curve import learning_curve

def plot_learning_curve(hidden_layer_sizes):
    train_sizes = np.linspace(0.2, 1, 20)
    N_train, val_train, val_test = learning_curve(MLPClassifier(hidden_layer_sizes),
                                                  X, y, train_sizes, cv=5, scoring=rms_error)
    plot_with_err(N_train, val_train, label='training scores')
    plot_with_err(N_train, val_test, label='validation scores')
    plt.xlabel('Training Set Size'); plt.ylabel('rms error')
    plt.ylim(0.1, 1)
    plt.xlim(0, 284)
    plt.legend()
```

# Summary

The Training Score shows how well a model fits the data it was trained on. This is not a good indication of model effectiveness
The Validation Score shows how well a model fits hold-out data. The most effective method is some form of cross-validation, where multiple hold-out sets are used.

1. Validation Curves are a plot of validation score and training score as a function of model complexity:
when the two curves are close, it indicates underfitting
when the two curves are separated, it indicates overfitting
the "sweet spot" is in the middle

2. Learning Curves are a plot of the validation score and training score as a function of Number of training samples:
when the curves are close, it indicates underfitting, and adding more data will not generally improve the estimator.
when the curves are far apart, it indicates overfitting, and adding more data may increase the effectiveness of the model.

These tools are powerful means of evaluating your model on your data.
